DATA VISUALIZATION:

Creating a histogram using foot_ball["Age"] column:
average(mean) = 25.789474
standard deviation(std) = 4.377354
total_rows(size) = 571


```python
data = np.random.normal(loc=25.789474, scale=4.377354, size=571)
plt.hist(data,bins=10,color='black', alpha=0.7)
plt.title("Histogram of Age Distribution Among All Players")
plt.ylabel("Frequency")
plt.xlabel("Age")
plt.show()
```

Creating a line graph of first 9 players' goals

```python
goals = [n for n in foot_ball["Goals"].head(9)]
names= [n for n in foot_ball["Name"].head(9)]

plt.figure(figsize=(16,6))
plt.plot(names, goals, marker='o', linestyle='-', color='black')
plt.xlabel("Names")
plt.ylabel("Goals")
plt.title("First 9 Players' Goals")
plt.grid(True)
plt.show()
```

Creating a Scatter Graph of Goals vs Appearances for the first 10 players

```python
goals = [n for n in foot_ball["Goals"].head(10)]
appearances= [n for n in foot_ball["Appearances"].head(10)]

plt.scatter(appearances,goals)
plt.title("Scatter Plot of Goals Scored vs Appearances")
plt.xlabel("Appearances")
plt.ylabel("Goals Scored")
plt.show()
```

Creating a Bar Graph for Appearances of the First Five Players

```python
players = [n for n in foot_ball["Name"].head()]
appearances = [n for n in foot_ball["Appearances"].head()]

plt.figure(figsize=(12, 5))
plt.bar(players, appearances, color='black', alpha=0.7)
plt.title('Game Appearances of each first five Players')
plt.xlabel('Players')
plt.ylabel('Appearances')
plt.show()
```

Creating a Pie Chart of the last Five Players' Goals

```python
Names = [n for n in foot_ball["Name"].tail()]
goals= [n for n in foot_ball["Goals"].tail()]

plt.figure(figsize=(7,7))
plt.pie(goals, labels=Names, autopct='%1.1f%%', startangle=15)
plt.title("Last 5 Players's Goals Scored")
plt.show()
```

Creating a Boxplot for Distribution of Yellow Cards

```python
Ycards = [n for n in foot_ball["Yellow cards"]]
plt.figure(figsize=(6, 8))
plt.boxplot(Ycards, vert=True, patch_artist=True, meanline=True, showmeans=True)
plt.title("Distribution of Yellow Cards")
plt.ylabel("Yellow Cards")
plt.grid(True)
plt.show()
```
